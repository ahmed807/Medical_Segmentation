```python
import torch
from diffusers import FluxFillPipeline
from diffusers.utils import load_image

# image = load_image("https://huggingface.co/datasets/diffusers/diffusers-images-docs/resolve/main/cup.png")
image = "/home/ahma/Medical_Segmentation/sam_finetuning_dataset/images/12_2_2.jpg"
# mask = load_image("https://huggingface.co/datasets/diffusers/diffusers-images-docs/resolve/main/cup_mask.png")
mask = "/home/ahma/unannotate/output/12/annotation_mask_2_2.png"
pipe = FluxFillPipeline.from_pretrained("black-forest-labs/FLUX.1-Fill-dev", torch_dtype=torch.bfloat16).to("cuda")
image = pipe(
    prompt="remove and fill",
    image=image,
    mask_image=mask,
    height=1632,
    width=1232,
    guidance_scale=30,
    num_inference_steps=50,
    max_sequence_length=512,
    generator=torch.Generator("cpu").manual_seed(0)
).images[0]
image.save(f"flux-fill-dev.png")
```

```python
import torch
from diffusers import FluxFillPipeline
from diffusers.utils import load_image
from PIL import Image

# Load images properly (PIL format required)
image = Image.open("/home/ahma/Medical_Segmentation/sam_finetuning_dataset/images/12_2_2.jpg").convert("RGB")
mask = Image.open("/home/ahma/unannotate/output/12/annotation_mask_2_2.png").convert("RGB")

# Load pipeline on CUDA device 0 explicitly
pipe = FluxFillPipeline.from_pretrained(
    "black-forest-labs/FLUX.1-Fill-dev",
    torch_dtype=torch.bfloat16
).to("cuda:0")

result = pipe(
    prompt="remove and fill",
    image=image,
    mask_image=mask,
    height=1632,
    width=1232,
    guidance_scale=30,
    num_inference_steps=50,
    max_sequence_length=512,
    generator=torch.Generator("cuda:0").manual_seed(0)  # generator on same device
).images[0]

result.save("flux-fill-dev.png")
```
